```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
stim_durs = [100, 150, 200, 250, 400, 700, 1000]
stim_durs = np.array(stim_durs)
```

```python
sigma_mean = np.load('sigma_mean.npy')
sigma_std = np.load('sigma_std.npy')
```

```python
sigma_mean
```

```python
plt.figure(figsize=(5,3))
plt.errorbar(1e2 / stim_durs[::-1], sigma_mean[::-1], sigma_std[::-1], marker='s', markersize=5, color='indianred')
plt.xlabel('Velocity [a.u.]')
plt.ylabel('Receptive filed size [a.u.]')
plt.title('Receptive field size depends on speed')
plt.tight_layout()
plt.savefig('velocity_dep.png')
```
